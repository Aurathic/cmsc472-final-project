```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
plt.style.use("ggplot")
```

```python
st = [  [0.355881, 0.621832, 0.738345, 0.794023],
        [0.3477869274318065, 0.6209469891919712, 0.7376479670612455, 0.7932321152856407],
        [0.345857, 0.621204, 0.738291, 0.793361],
        [0.349846, 0.617730, 0.732887, 0.790015],
        [0.347787, 0.600617, 0.705095, 0.760165]]
batch_sizes = [16, 64, 128, 1024, 4096]
accs_batch_sizes = [*zip(*st)]
print(accs_batch_sizes)
```

```python
plt.figure(figsize=(7,5))
fig1 = plt.gcf()
colors = ['tab:red', 'tab:orange', 'tab:green', 'tab:blue']
for i, (xs, c) in enumerate(zip(accs_batch_sizes, colors)):
  plt.plot(batch_sizes, xs, color=c, marker='o', label=f"top-{i+1}")
plt.xlabel("batch size"); plt.ylabel("accuracy", )
plt.xscale('log')
plt.legend(loc='best', bbox_to_anchor=(0.2,0.5))
plt.show()
```
