```python
file_name = "ensemble_eval-4.csv"
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(file_name)
```

```python
keys = [' Allergy / Immunology', ' Autopsy', ' Bariatrics', ' Cardiovascular / Pulmonary', ' Chiropractic', ' Consult - History and Phy.', ' Cosmetic / Plastic Surgery', ' Dentistry', ' Dermatology', ' Diets and Nutritions', ' Discharge Summary', ' Emergency Room Reports', ' Endocrinology', ' ENT - Otolaryngology', ' Gastroenterology', ' General Medicine', ' Hematology - Oncology', ' Hospice - Palliative Care', ' IME-QME-Work Comp etc.', ' Lab Medicine - Pathology', ' Letters', ' Nephrology', ' Neurology', ' Neurosurgery', ' Obstetrics / Gynecology', ' Office Notes', ' Ophthalmology', ' Orthopedic', ' Pain Management', ' Pediatrics - Neonatal', ' Physical Medicine - Rehab', ' Podiatry', ' Psychiatry / Psychology', ' Radiology', ' Rheumatology', ' Sleep Medicine', ' SOAP / Chart / Progress Notes', ' Speech - Language', ' Surgery', ' Urology']
```

```python
df
```

```python
df['label_name'] = df['label'].apply(lambda i: keys[i])
```

```python
# returns indices of top_num number of elements in list l
def argmax(l, top_num=1):
    top = []
    for t in range(top_num):
        index = None
        max = -100
        for i in range(len(l)):
            if l[i] > max:
                index = i
                max = l[i]
        l.remove(max)
        top.append(index)

    return top
```

```python
# used to convert output col in csv to array of floats
def string_to_list(x):
    s = x[2:len(x)-2].strip()
    s = s.replace('\n  ', ' ')
    s = s.split()
    s = [float(i) for i in s]
    return s
```

Add column for top 4 predictions

```python
df['output_top_4'] = df['output'].apply(lambda x: argmax(string_to_list(x), top_num=4))
```

Add column converting indices to category name

```python
df['top_4_labels'] = df['output_top_4'].apply(lambda x: [keys[i] for i in x])
```

```python
clean_df = df[['text', 'label_name', 'top_4_labels']]
```
